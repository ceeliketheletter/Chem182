```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("I2_SlowScan_ForReals_PeakPick.txt", skip_header = 2, delimiter = ',')
number = data[:,0]
wavelength = data[:,2]
Abs = data[:,3]
```

```python
#plt.figure(figsize=(15,10))
plt.plot(number, wavelength)
plt.xlabel("Number")
plt.ylabel("Wavelength")
plt.title("I2_SlowScan_ForReals")
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(wavelength, Abs)
plt.xlabel("Wavelength [nm]")
plt.ylabel("Abs")
plt.show()
```
